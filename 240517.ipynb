# **Seaborn**

# 1) scatterplot

```python
iris_data = pd.read_csv('iris.csv')
sns.scatterplot(data=iris_data, x='petal_length', y='petal_width', hue='species')
plt.title('Scatter Plot by seaborn',fontsize=20)
plt.show()
```

```python
data = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})
data
```

```python
sns.scatterplot(data=data, x='x', y='y', hue='category')
plt.show()
```

# 2) countplot

```python
sns.countplot(x='species', data=iris_data)
plt.show()
```

'category'          

```python
data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})
data
```

```python
sns.countplot(x='category', data=data,hue='category')
plt.show()
```

```python
penguins_data = sns.load_dataset('penguins')
sns.barplot(x='species', y='body_mass_g', data=penguins_data,hue='sex')
plt.show()
```
