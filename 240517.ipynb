```python
data1 = {'도시':['서울','부산','인천','대구'],'인구 수':[9904312,3448737,2890451,2466052]}
df2 = pd.DataFrame(data1)
df2
```

도시별 인구 수를 수직 막대 그래프로 시각화하세요

```python
plt.bar(df2['도시'],df2['인구 수'])
plt.show()
```

그래프의 제목을 도시별 인구 수로 설정하세요

# **Seaborn**

# 1) scatterplot

```python
iris_data = pd.read_csv('iris.csv')
sns.scatterplot(data=iris_data, x='petal_length', y='petal_width', hue='species')
plt.title('Scatter Plot by seaborn',fontsize=20)
plt.show()
```

실습문제

```python
data = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})
data
```

```python
sns.scatterplot(data=data, x='x', y='y', hue='category')
plt.show()
```

# 2) countplot

```python
sns.countplot(x='species', data=iris_data)
plt.show()
```

실습문제

'category' 열을 사용하여 각 카테고리별로 데이터의 개수를 세어서 막대 그래프로 시각화하세요

```python
data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})
data
```

```python
sns.countplot(x='category', data=data,hue='category')
plt.show()
```
